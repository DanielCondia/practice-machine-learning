```python
import pandas as pd
```

```python
# load data from csv
data_csv = pd.read_csv('./csv/data_csv.csv')
# read header of csv
data_csv.head()
```

```python
# review data info
data_csv.info()
```

```python
# find and handle missing values
data_csv.isnull().sum()
```

```python
# handle missing values
data_csv['goles_local'] = data_csv['goles_local'].fillna(data_csv['goles_local'].mean())
print("Valores nulos antes de imputación: ")
print(data_csv.isnull().sum())
```

```python
prepare_data = pd.get_dummies(data_csv, columns=['equipo_local', 'equipo_visitante'])
prepare_data.head()
```

```python
# Drop data duplicates
print("rows before removing duplicates: ", len(prepare_data))
prepare_data.drop_duplicates(inplace=True)
print("rows after removing duplicates: ", len(prepare_data))
```

```python
# handle dates
prepare_data['fecha_partido'] = pd.to_datetime(prepare_data['fecha_partido'], errors='coerce')

print('invalid dates (NaT) after conversion: ')
print(prepare_data['fecha_partido'].isnull().sum())
```
