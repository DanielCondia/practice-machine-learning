```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load data from csv
df_csv = pd.read_csv('./csv/data_csv.csv')
df_csv.head(10)
```

```python
# general info
df_csv.describe()
```

```python
# average goals of goals scored at home and away
goals_home = df_csv[df_csv['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goals_away = df_csv[df_csv['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print('Average goals scored at home: ', goals_home)
print('Average goals scored away: ', goals_away)
```

```python
# histogram of goals scored at home and away
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df_csv['goles_local'], kde=True, ax=ax[0], bins=10)
ax[0].set_title('Goals scored home')

sns.histplot(df_csv['goles_visitante'], kde=True, ax=ax[1], bins=10, color='orange')
ax[1].set_title('Goals scored away')

plt.show()
```

```python
# boxplot of goals scored at home and away
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.boxplot(x=df_csv['goles_local'], ax=ax[0])
ax[0].set_title('Goals scored home')

sns.boxplot(x=df_csv['goles_visitante'], ax=ax[1], color='orange')
ax[1].set_title('Goals scored away')

plt.show()
```

```python
# scatterplot of possession vs goals scored at home
sns.scatterplot(x='posesion_local (%)', y='goles_local', data=df_csv)
plt.title('Relationship between possession and goals scored at home')
plt.show()
```

```python
# heat map for correlations of key variables
plt.figure(figsize=(10, 6))
sns.heatmap(df_csv[['goles_local', 'goles_visitante', 'posesion_local (%)', 'posesion_visitante (%)',
                    'tiros_arco_local', 'tiros_arco_visitante']].corr(), annot=True, cmap='coolwarm')
plt.title('correlations heatmap')
plt.show()
```
